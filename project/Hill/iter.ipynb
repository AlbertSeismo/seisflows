```python
import obspy
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True
import numpy as np
import pandas as pd
```

```python
f = open('./output.optim','r')
lines = len(f.readlines()) 
f.close() 

f = open('./output.optim','r')
f.readline()
f.readline()
#f.readline()
data = []
for _ in range(lines-2):
    line = f.readline()
#        print line[0:10] != '          '
    if line[0:10] != '          ' and line != '\n':
        data.append(line[:-1])
f.close()       
```

```python
ITER = []
MISFIT = []
for i in range(len(data)):
    ITER.append(int(data[i][0:10]))
    MISFIT.append(float(data[i][25:37]))
```

```python
plt.figure(figsize=(10, 4), dpi=200)
plt.plot(ITER, np.array(MISFIT)/MISFIT[0], linewidth = 3)
plt.tick_params(labelsize=15)
plt.ylabel(r'$normalized$ $model$ $misfit$', fontsize = 20)
plt.xlabel(r'$iteration$', fontsize = 20)
plt.ylim(-0.1, 1.1)
plt.xlim(-1, 100)
plt.grid(True)
plt.tight_layout()
plt.savefig('./hill.jpg', format='png')
```
